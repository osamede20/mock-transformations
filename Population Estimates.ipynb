### Requirements

#### Observations & Dimensions

The `observations` should be apparent.

The required dimensions are:

* **Geography** - get this from the top row "title", cut off the non place names.
* **Time** - use dates from left hand column
* **CDID** - from across the top

````python
from databaker.framework import *
import pandas as pd

# the Seasonally Adjusted Tabs
tabs = loadxlstabs("./sources/pop.xls")
````

````python
tab = tabs[0]
country = tab.excel_ref('B1').expand(RIGHT).is_not_blank()
year = tab.excel_ref('A8').expand(DOWN).is_not_blank()
cdid = tab.excel_ref("B2").expand(RIGHT).is_not_blank()
observations = year.fill(RIGHT).is_not_blank()

dimensions = [
    HDim(year, 'Year', DIRECTLY, LEFT),
    HDim(country, "Geography", DIRECTLY, ABOVE),
    HDim(cdid, "CDID", DIRECTLY, ABOVE)
    ]
cs= ConversionSegment(tab, dimensions, observations)
datacube = cs.topandas()
````

````python
datacube
````

````python
datacube["Geography"] = datacube["Geography"].str\
.replace("population mid-year estimate", " ")

datacube["CDID"] = datacube["CDID"].str.rstrip("OP")

datacube["OBS"] = datacube["OBS"].astype(int)
````
